```python
escolas = pd.read_csv(url, delimiter=';')
sb.displot(escolas['ANOS FINAIS'], bins=10)
plt.pyplot.show()
```

```python
escolas = pd.read_csv(url, delimiter=';')
sb.displot(escolas['ANOS INICIAIS'], bins=10)
plt.pyplot.show()
```

```python
filtro_escola = escolas[escolas['ANOS INICIAIS'] > 0]
sb.displot(filtro_escola['ANOS INICIAIS'], bins=10)
plt.pyplot.show()
```

```python
filtro_escola = escolas[escolas['ENSINO MEDIO'] > 0]
sb.displot(filtro_escola['ENSINO MEDIO'], bins = 10)
plt.pyplot.show()
```

```python
fundamental_inicio = escolas['ANOS INICIAIS'].sum()
fundamental_final = escolas['ANOS FINAIS'].sum()
medio = escolas['ENSINO MEDIO'].sum()
(fundamental_inicio + fundamental_final + medio)
```

```python
alunos = [fundamental_inicio, fundamental_final, medio]
periodo = ['fundamental_inicio', 'fundamental_final', 'medio']
plt.pyplot.pie(alunos, labels=periodo, autopct='%0.0f%%')
plt.pyplot.show()
```

```python
# Media relativa

filtro_escola = escolas[escolas['ENSINO MEDIO'] > 0]
sb.boxenplot(y=filtro_escola['ENSINO MEDIO'])
plt.pyplot.show()
```

```python
# Gerar uma descrição estatística da coluna 'ANOS_FINAIS'

escolas['ANOS FINAIS'].describe()
```

```python
# Gráfico de dispersão

filtro_escola = escolas[(escolas['ANOS INICIAIS'] > 0) & (escolas['ANOS FINAIS']>0)]
sb.scatterplot(x=filtro_escola['ANOS INICIAIS'], y=filtro_escola['ANOS FINAIS'])
plt.pyplot.show()
```
