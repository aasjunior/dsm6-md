```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

sb.set(rc={'figure.figsize':(15,8)})
```

```python
url = r'C:\Users\isabelle.hondo\Downloads\conjunto-dados.csv'
ibge = pd.read_csv(url, delimiter=';')
ibge.head(20)
```

```python
ibge.info()
```

```python
print(ibge.describe())
```

```python
# Descobrir valores únicos de uma coluna
print(ibge['Quantidade Recursos'].unique())

# Contar a frequência de valores em uma coluna
print(ibge['Quantidade Recursos'].value_counts())
```

```python
# Verificar valores vazios
print(ibge.isnull().sum())
```

```python
colunas_desejadas = ibge[['Quantidade Downloads', 'Quantidade Recursos']]
print('\nDados das colunas Quantidade Downloads e Quantidade Recursos:')
print(colunas_desejadas)
```

```python
# Consulta a uma linha específica

indice = 3
linha_desejada = ibge.iloc[indice]

print('\nDados da linha', indice, ':')
print(linha_desejada)
```

```python
# Acesso a uma linha e coluna específica
linha_indice = 2
coluna_nome = 'Quantidade Seguidores'
celula_desejada = ibge.loc[linha_indice, coluna_nome]
print('\nDado na Linha: ', linha_indice, 'Dado na coluna:', coluna_nome, ':')
print(celula_desejada)
```

```python
# Graficos histograma
colunas_desejadas = ibge[['Quantidade Downloads', 'Quantidade Recursos']]
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(colunas_desejadas['Quantidade Downloads'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Distribuição de Downloads')
plt.xlabel('Quantidade Downloads')
plt.ylabel('Frequência')

```

```python
plt.subplot(1, 2, 1)
plt.hist(colunas_desejadas['Quantidade Recursos'].dropna(), bins=20, color='salmon', edgecolor='black')
plt.title('Distribuição de rexursos')
plt.xlabel('Quantidade recusros')
plt.ylabel('Frequência')
```
